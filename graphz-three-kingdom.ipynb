```python
from graphviz import Digraph
```

```python
dot = Digraph('Three-Kingdoms', filename='Three-Kingdoms.gv')
```

```python
dot.node('A', 'Wu')

dot.node('B', 'Wei')

dot.node('C', 'Shu')

dot.node('D', 'Jin')
```

```python
dot.edges(['AD', 'BD', 'CD'])
```

```python
print(dot.source)
```

```python
dot.render('Three-Kingdoms.gv', view=True)  
```
